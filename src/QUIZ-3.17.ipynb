```kotlin
val spices = listOf("curry", "pepper", "cayenne", "ginger", "red curry", "green curry", "red pepper")

spices
    .filter { it.contains("curry") }
    .sortedBy { it.length }
```

```kotlin
spices.filter { it.startsWith("c") && it.endsWith("e") }
```

```kotlin
spices.take(3).filter { it.startsWith("c") }
```
